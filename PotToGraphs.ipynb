```sagemath
import numpy as np
import scipy
import sys
import copy
import time 
import itertools
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'] 

#PART 0: User Input

pot_str = 'a3,b;-a2,b,-b;b,-b3' #?if you wish to copy paste in as one string (from a spreadsheet or where ever), seperate each bond type in tile with a comma and each tile with a semi-colon
order = 0 #? Options for order: 0 if unsure of min order, negative number if want only graphs of specified order only, 
            #? or use any order greater than min order if want all graphs up to that order to be generated (if given order is too small, will default to min order
choose_tile_ratios = False # Choose if user wants to specified ratios to be used to generate graphs, user will be prompted once ratios have been generated #!(manual inputs are implemented atm)
non_iso_graphs = False #Allows all non-isomorphic iterations of the graph ratios NOTE: This will significantly increase graph production -- allow several minutes for generation, time can be decreased by choosing fewer ratios
avoid_loops = True # Loops may occur if this is True, but are not prioritized
avoid_multiple_edges = True # Multiple edges may occur if this is True, but are not prioritized

max_graphs = 100 #input max number of graphs to be produced (this does not include non-iso iterations, for now ...)

inital_time1 = time.perf_counter()

#PART 1: Matrix Construction
def convert_pot(pot_str):
    def str_to_list(pot_str):
        if type(pot_str) == list:
            num_tiles = len(pot_str)
            return(pot_str)
        else:
            pot_str = pot_str.replace(" ","")
            str_pot = pot_str.split(";")
            num_tiles = len(str_pot)
        return(str_pot,num_tiles)  
    str_pot, num_tiles = str_to_list(pot_str)

    def B(str_pot):
        unique_bond_edge_types = {'a'}
        for tile in str_pot:
            #break the tile into pieces
            edges = tile.split(',')
            #check each piece
            for edge in edges:
                #if its negative, go past
                if edge[0] == '-':
                    unique_bond_edge_types.add(edge[1])
                else:
                    unique_bond_edge_types.add(edge[0])
        return (len(unique_bond_edge_types))
    num_bond_edge_types = B(str_pot)

    def pot_matrix(num_bond_edge_types, num_tiles, str_pot):
        out_mat = [([0]*(num_tiles)) for j in range(num_bond_edge_types)]
        for column in range(len(str_pot)):
            tile = str_pot[column]
            edges = tile.split(',')
            for edge in edges:
                #if we have a long tile with a leading -, we must be a hatted char
                if len(edge) >= 3 and edge[0] == '-':
                    out_mat[ord(edge[1])-97][column] -= int(edge[2:])
                #if we have a long tile without a leading -, we must be an unhatted char
                elif len(edge) >= 2 and edge[0] != '-':
                    out_mat[ord(edge[0])-97][column] += int(edge[1:])
                #if we arent long, we should check if we are negative
                elif edge[0] == '-':
                    out_mat[ord(edge[1])-97][column] -= 1
                elif (edge[0]).isalpha():
                    out_mat[ord(edge[0])-97][column] += 1
        return(out_mat)

    def pot_vector(num_bond_edge_types, num_tiles, str_pot):
        vector_pot = [([0]*(num_bond_edge_types*2)) for i in range(num_tiles)]
        for row in range(num_tiles):
            tile = str_pot[row]
            edges1 = tile.split(',')
            for edge1 in edges1:
                if len(edge1) >= 3 and edge1[0] == '-':
                    vector_pot[row][(ord(edge1[1])-97)+1+(ord(edge1[1])-97)] += int(edge1[2:])
                elif len(edge1) >= 2 and edge1[0] != '-':
                    vector_pot[row][ord(edge1[0])-97+(ord(edge1[0])-97)] += int(edge1[1:])
                elif edge1[0] == '-':
                    vector_pot[row][(ord(edge1[1])-97)+1+(ord(edge1[1])-97)] += 1
                elif (edge1[0]).isalpha():
                    vector_pot[row][ord(edge1[0])-97+(ord(edge1[0])-97)] += 1
        
        return(vector_pot)

    return (pot_matrix(num_bond_edge_types, num_tiles, str_pot), pot_vector(num_bond_edge_types, num_tiles, str_pot))
M, tiles = convert_pot(pot_str) #pot matrix, vector matrix
M_test = copy.deepcopy(M) #copy of pot matrix to check order
print(M, tiles)

#PART 2: Pot evaluation
    #PART 2.1: Minimum order check
def check_max_order(M):
    for index in range(len(M)):
        # If the matrix is not rectangular, that means it was entered correctly
        if(len(M[index]) != len(M[0])):
            print("Matrix is not rectangular, row 0 is size " + str(len(M[0])) + " and row " + str(index) + " is length " + str(len(M[index])))
            sys.exit()
        # If a row has no negative values there's no 'hatted' bond edges of that type, indicating an input error
        if(min(M[index]) >= 0):
            print("Row " + str(index) + " has no negative values, so a graph can't be mate")
            sys.exit()
        # # If a row has no positive values there's no 'non-hatted' bond edges of that type, indicating an input error
        if(max(M[index]) <= 0):
            print("Row " + str(index) + " has no positive values, so a graph can't be mate")
            sys.exit()
    # Build the various vectors for our integer linear program
    # The c vector is all 1s, to minimuze the sum of tiles
    c = [1 for i in M[0]]
    # The constraint matrix is just our construction matrix
    a = M
    # Lower and upper bounds for each constraint are zero (to satisfy the sum of each bond edge type is 0)
    b_l = [0 for i in a]
    b_u = [0 for i in a]
    # Add an extra constraint, ensuring that the sum of tiles we use is 1 or greater (to make sure we don't just generate the empty graph)
    a.append([1 for i in c])
    b_l.append(1)
    b_u.append(999999) #Arbitrarily high upper bound
    # This constraint forces each decision variable to have an integer value, since we can't have decimal values of tiles
    integrality = [1 for i in c]
    # Plug our constraints into scipy's mixed-integer linear programming algorithm
    constraints = scipy.optimize.LinearConstraint(a, b_l, b_u)
    res = scipy.optimize.milp(
        c=c,
        constraints=constraints,
        integrality=integrality,
        options={"disp": False})

    # Check the status of the result and exit the program if an error happened
    # if(int(res.status) == 0):
    #     print("Optimal Solution Found")
    if(int(res.status) == 1):
        print("Operation Timed Out")
        sys.exit()
    elif(int(res.status) == 2):
        print("Problem is infeasible. Check construction matrix input.")
        sys.exit()
    elif(int(res.status) == 3):
        print("Problem is unbounded. Check construction matrix input.")
        sys.exit()
    elif(int(res.status) == 4):
        print("Unknown error.")
        sys.exit()

    # print("Amounts of each tile type to build the graph: " + str(res.x))
    # print("Minimum size of graph constructed from pot: " + str(res.fun))
    return(int(res.fun))
try_order = check_max_order(M_test)
specific_order = False
if order < 0:
    specific_order = True
    order = abs(order)
if try_order > order:
    order = try_order
    print('Input order not viable, order for graph generation has defaulted to,', order)
 
    #PART 2.2: Pot ratio construction
def pot_ratios(M, order):
    def find_all_optimal_solutions_from_pot_matrix(pot_matrix, max_order):
        A, lb, ub = buildConstraints(pot_matrix, max_order)
        return find_optimal_solutions(A, lb, ub)

    def buildConstraints(cons_matrix, max_order):
        # Build the various vectors for our integer linear program
        # The c vector is all 1s, to minimuze the sum of tiles
        c = [1 for i in cons_matrix[0]]
        # The constraint matrix is just our construction matrix
        a = cons_matrix
        # Lower and upper bounds for each constraint are zero (to satisfy the sum of each bond edge type is 0)
        b_l = [0 for i in a]
        b_u = [0 for i in a]
        # Add an extra constraint, ensuring that the sum of tiles we use is 1 or greater (to make sure we don't just generate the empty graph)
        a.append([1 for i in c])
        b_l.append(1)
        b_u.append(max_order) # arbitrarily high upper bound
        return np.array(a), np.array(b_l).reshape((len(b_l), 1)), np.array(b_u).reshape((len(b_u), 1))

    def find_optimal_solutions(A, lb, ub):
    # # finds all optimal solutions x satisfying lb <= Ax <= ub, where lb_x <= x <= ub_x
        U = np.hstack((ub, -A))
        L = np.hstack((-lb, A))
        dt = np.zeros((A.shape[1],1))
        lower_bound_x = np.hstack((dt, np.eye(A.shape[1])))
        #upper_bound_x = np.hstack((ub_x, -np.eye(A.shape[1])))
        T = np.vstack((U, L, lower_bound_x)).astype('int')
        #T = np.vstack((U, L)).astype('int')
        #just_b = np.vstack((lower_bound_x, upper_bound_x)).astype('int')
        K = Polyhedron(ieqs=list(T))
        return K.integral_points()

    ratios = [list(ele) for ele in (find_all_optimal_solutions_from_pot_matrix(M, order))]
    return(ratios)
    # return(find_all_optimal_solutions_from_pot_matrix(M, order))
maybe_possible_multiplicities = (pot_ratios(M, order))
possible_multiplicities = []
if specific_order:
    # max_created_order = sum(maybe_possible_multiplicities[len(maybe_possible_multiplicities)-1])
    max_created_order = 0
    for i in range(len(maybe_possible_multiplicities)):
        sums = sum(maybe_possible_multiplicities[i])
        if sums == order:
            max_created_order = order
            break
        elif sums > max_created_order:
            max_created_order = sums
    if max_created_order < order:
        order = max_created_order
        print("Ratios could not be created for input order, defaulted to max order in range of input: ", order)
    for i in range(len(maybe_possible_multiplicities)):
        if sum(maybe_possible_multiplicities[i]) == order:
            possible_multiplicities.append(maybe_possible_multiplicities[i])
else:
    possible_multiplicities = maybe_possible_multiplicities

temp_time = 0
if choose_tile_ratios:
    temp_time = time.perf_counter()
    print('To pick ratios to generate graphs from, input a list of keys (ex: 0,1,3,5) of the ratio dictionary below:')
    mult_dict1 = {}
    for i, name in enumerate(possible_multiplicities,1):
        mult_dict1[i] = name
    print(mult_dict1)
    chosen = []
    ask = (input(': '))
    ask = ask.split(',')
    ask = [eval(i) for i in ask]
    for i in range(len(ask)):
        chosen.append(possible_multiplicities[ask[i]])
    possible_multiplicities = chosen
    print('Number of graphs being generated:', len(possible_multiplicities))
    if len(possible_multiplicities) > max_graphs:
        cont = input('Maximum graphs limit to be exceeded by output, would you still like to continue? (Y/N)')
        if cont == 'N':
            print('Terminating program...')
            sys.exit()
    print("Chosen ratio(s): ", possible_multiplicities)
    temp_time1 = time.perf_counter()
    temp_time = temp_time1 - temp_time
else:
    mult_dict = {}
    for i, name in enumerate(possible_multiplicities,1):
        mult_dict[i] = name
    print('Number of graphs being generated:', len(mult_dict))
    if len(possible_multiplicities) > max_graphs:
        temp_time = time.perf_counter()
        cont = input('Maximum graphs limit to be exceeded by output, would you still like to continue? (Y/N)')
        if cont == 'N':
            print('Terminating program...')
            sys.exit()
        temp_time1 = time.perf_counter()
        temp_time = temp_time1 - temp_time
    print(mult_dict)

part_one_two_time = time.perf_counter() - inital_time1 - temp_time

#PART 3: Graph construction
# Return an ordering where the first ordering is further sorted by a second list of values
def compound_ordering_with_values(ordering_1, ordering_1_values, ordering_2_values):
    lists = [[]]
    list_counter = 0
    for i in range(len(ordering_1) - 1):
        this_rank = ordering_1[i]
        next_rank = ordering_1[i + 1]
        lists[list_counter].append(this_rank)
        if ordering_1_values[this_rank] != ordering_1_values[next_rank]:
            list_counter += 1
            lists.append([])
    lists[len(lists) - 1].append(ordering_1[len(ordering_1) - 1])
    for i in range(len(lists)):
        lists[i] = sorted(lists[i], key=lambda tile: ordering_2_values[tile])
    ordering = sum(lists, [])
    return ordering

# Return an ordering where the first ordering is further sorted by a second ordering
def compound_ordering_with_ordering(ordering_1, ordering_1_values, ordering_2):
    lists = [[]]
    list_counter = 0
    for i in range(len(ordering_1) - 1):
        this_rank = ordering_1[i]
        next_rank = ordering_1[i + 1]
        lists[list_counter].append(this_rank)
        if ordering_1_values[this_rank] != ordering_1_values[next_rank]:
            list_counter += 1
            lists.append([])
    lists[len(lists) - 1].append(ordering_1[len(ordering_1) - 1])
    for i in range(len(lists)):
        lists[i] = sorted(lists[i], key=lambda tile: ordering_2.index(tile))
    ordering = sum(lists, [])
    return ordering

# Return a list of lists which are sorted by the first ordering and then the second ordering. each sublist contains indices of things that have the same ordering 1 value
def compound_ordering_with_ordering_list(ordering_1, ordering_1_values, ordering_2):
    lists = [[]]
    list_counter = 0
    for i in range(len(ordering_1) - 1):
        this_rank = ordering_1[i]
        next_rank = ordering_1[i + 1]
        lists[list_counter].append(this_rank)
        if ordering_1_values[this_rank] != ordering_1_values[next_rank]:
            list_counter += 1
            lists.append([])
    lists[len(lists) - 1].append(ordering_1[len(ordering_1) - 1])
    for i in range(len(lists)):
        lists[i] = sorted(lists[i], key=lambda tile: ordering_2.index(tile))
    ordering = sum(lists, [])
    return lists

def list_of_tuples(l1, l2):
    return list(map(lambda x, y:(x,y), l1, l2))

degrees = []
diversities = []

# Find the degree of each tile and store it in a list
for tile in tiles:
    degrees.append(sum(tile))

# Find the diversity of each tile and store it
for tile in tiles:
    diversity = 0
    for exponent in tile:
        if exponent != 0:
            diversity += 1
    diversities.append(diversity)

# Sort by degree and diversity
degrees_ordering = sorted(range(len(tiles)), key=lambda tile: -degrees[tile]) #store the order we should go through tiles according to their degree. change minus sign to go from high to low
diversities_ordering = sorted(range(len(tiles)), key=lambda tile: diversities[tile]) #store the order we should go through tiles according to their diversity

# Sort alphabetically
def alphabetical_step(lists, i):
    final_alphabetical = []
    if i < len(tiles[0]) - 1:
        for j in range(len(lists)):
            if len(lists[j]) == 1:
                final_alphabetical.append(lists[j][0])
            else:
                first_col_ordering = sorted(lists[j], key=lambda tile: -tiles[tile][i])
                second_col_ordering = sorted(lists[j], key=lambda tile: -tiles[tile][i+1])
                sublists = compound_ordering_with_ordering_list(first_col_ordering,[tiles[k][i] for k in range(len(tiles))], second_col_ordering)
                final_alphabetical += alphabetical_step(sublists, i+1)
    else:
        temp = []
        for j in range(len(lists)):
            for k in lists[j]:
                temp.append(k)
        return temp
    return final_alphabetical

first_col_ordering = sorted(range(len(tiles)), key=lambda tile: -tiles[tile][0])
second_col_ordering = sorted(range(len(tiles)), key=lambda tile: -tiles[tile][1])
numbered_lists = compound_ordering_with_ordering_list(first_col_ordering,[tiles[k][0] for k in range(len(tiles))], second_col_ordering)
alphabetical_ordering = alphabetical_step(numbered_lists, 1)

# Find an ordering based on priority of degree then diversity
degrees_diversities_ordering = compound_ordering_with_values(degrees_ordering, degrees, diversities)

# Sort compound ordering alphabetically
degrees_diversities_tuple = list_of_tuples(degrees, diversities)
degrees_diversities_alphabetical_ordering = compound_ordering_with_ordering(degrees_diversities_ordering, degrees_diversities_tuple, alphabetical_ordering)

# Returns a bond dictionary and the ratio used to construct it and a list of tile types in order of the vertex labels
def construct_bond_dictionary(tiles, multiplicities, ordering, avoid_multiple_edges):
    # Initialize bond dictinary to have as many keys and bond-edge types
    bond_dictionary = {}
    for i in range(len(tiles[0])/2):
        bond_dictionary.update({i+1:[]})

    # As we make connections, bond-edges will be deleted from this list
    tiles_not_yet_used = []

    # As we add new tiles into the graph, their labels will be added to this list
    tile_labels = []

    # Create graphs for each tile, ordered by the inputed ordering
    for i in ordering:
        for j in range(multiplicities[i]):
            tiles_not_yet_used.append(tiles[i].copy())
    
    tiles_original = tiles_not_yet_used.copy()

    # Prepare list to remember which vertex each tile is centered at in the graph
    tile_locations = []
    for i in range(len(tiles_not_yet_used)):
        tile_locations.append(-1)

    # Start adding tiles to the graph
    tile_locations[0] = 0
    tile_labels.append(tiles.index(tiles_original[0]))
    tile_in_focus = 0

    loops_allowed_now = not avoid_loops

    def update_locations(tile_locations, tile_new):
        new_tile_locations = tile_locations
        if tile_locations[tile_new] == -1:
            tiles_in_graph = 0
            for i in tile_locations:
                if i != -1:
                    tiles_in_graph += 1
            new_tile_locations[tile_new] = tiles_in_graph
            tile_labels.append(tiles.index(tiles_original[tile_new]))
        return new_tile_locations

    def connect_bond_edges(bond_dictionary, tile_in_graph, tile_new, bond_edge_type, bond_edge_index, match_index, tiles_not_yet_used):
        # print("Before:")
        # print(bond_dictionary)
        tiles_not_yet_used[tile_in_graph][bond_edge_index] -= 1
        tiles_not_yet_used[tile_new][match_index] -= 1
        if bond_edge_type > 0:
            bond_dictionary[bond_edge_type].append([tile_locations[tile_in_graph], tile_locations[tile_new]])
        else:
            bond_dictionary[-bond_edge_type].append([tile_locations[tile_new], tile_locations[tile_in_graph]])
        # print("After:")
        # print(bond_dictionary)

    # Avoiding multiple edges by matching one bond type at once between a pair of tiles
    if avoid_multiple_edges:
        done = False
        while not done:
            match_found = False
            for i in range(0, len(tiles_not_yet_used)):
                if not loops_allowed_now:
                    if i == tile_in_focus: # This prevents loops in the graph
                        continue
                bond_edge_type = 0 # Start by searching for a match to this bond_edge_type
                bond_edge_index = 0
                match_index = 0
                # Save first unconnected bond type found on the tile in focus
                for j,exponent in enumerate(tiles_not_yet_used[tile_in_focus]):
                    if exponent > 0:
                        bond_edge_index = j
                        bond_edge_type = ((-1)**j)*floor((j+2)/2)
                        if bond_edge_type > 0:
                            match_index = j+1
                        else:
                            match_index = j-1
                        break
                if bond_edge_type == 0: # i.e. if all exponents in this tile were 0, so the tile is fully used, then we can stop this loop
                    match_found = True
                    if (tile_in_focus < len(tiles_not_yet_used) - 1):
                        tile_in_focus += 1
                        tile_locations = update_locations(tile_locations, tile_in_focus)
                    else:
                        done = True
                    break
                if tiles_not_yet_used[i][match_index] > 0:
                    match_found = True
                    tile_locations = update_locations(tile_locations, i)
                    connect_bond_edges(bond_dictionary, tile_in_focus, i, bond_edge_type, bond_edge_index, match_index, tiles_not_yet_used) 
            if not match_found:
                loops_allowed_now = True
        return [bond_dictionary, multiplicities, tile_labels]
    
    # Creating multiple edges by matching many bond types at once between a pair of tiles
    else:
        done = False
        while not done:
            i = 0
            match_found = False
            while i < len(tiles_not_yet_used):
                if not loops_allowed_now:
                    if i == tile_in_focus: # This prevents loops in the graph
                        i += 1
                        continue
                exhausted = False
                while not exhausted:                
                    types = []
                    bond_edge_type = 0 # Start by searching for a match to this bond_edge_type
                    bond_edge_index = 0
                    match_index = 0
                    # Save all the unconnected bond edge types found on tile in focus
                    for j,exponent in enumerate(tiles_not_yet_used[tile_in_focus]):
                        if exponent > 0:
                            bond_edge_index = j
                            bond_edge_type = ((-1)**j)*floor((j+2)/2)
                            if bond_edge_type > 0:
                                match_index = j+1
                            else:
                                match_index = j-1
                            types.append([bond_edge_type, bond_edge_index, match_index])
                    if types == []: # i.e. if all exponents in this tile were 0, so the tile is fully used, then we can stop this loop
                        match_found = True
                        if (tile_in_focus < len(tiles_not_yet_used) - 1):
                            tile_in_focus += 1
                            tile_locations = update_locations(tile_locations, tile_in_focus)
                        else:
                            done = True
                        i = len(tiles_not_yet_used)
                        break
                    exhausted = True
                    for t in types:
                        bond_edge_type = t[0] # Start by searching for a match to this bond_edge_type
                        bond_edge_index = t[1]
                        match_index = t[2]
                        
                        if tiles_not_yet_used[i][match_index] > 0:
                            exhausted = False
                            match_found = True
                            tile_locations = update_locations(tile_locations, i)
                            connect_bond_edges(bond_dictionary, tile_in_focus, i, bond_edge_type, bond_edge_index, match_index, tiles_not_yet_used) 
                i += 1
            if not match_found:
                loops_allowed_now = True
        return [bond_dictionary, multiplicities, tile_labels]

def graph_from_bdict(bond_dictionary):
    graph = Graph()
    graph.allow_multiple_edges(True)
    graph.allow_loops(True)

    for i in bond_dictionary.keys():
        value = bond_dictionary[i]
        if value != None:
            graph.add_edges(value)

    return graph

def labeled_graph_from_bond_dictionary(bond_dictionary, labels):
    graph = DiGraph()
    graph.allow_multiple_edges(True)
    graph.allow_loops(True)

    for i in bond_dictionary.keys():
        value = bond_dictionary[i]
        if value != None:
            for j in value:
                graph.add_vertex(j[0])
                graph.set_vertex(j[0], '$t_{' + str(labels[j[0]]+1) + '}$')
                graph.add_vertex(j[1])
                graph.set_vertex(j[1], '$t_{' + str(labels[j[1]]+1) + '}$')
                graph.add_edge(j[0], j[1], '$' + alphabet[i-1] + '$')

    return graph

#PART 4: Checking for non-isomorphic graphs

# Returns (non_iso_list, iso_list), lists of bond dictionaries for all non-isomorphic graphs to the main input graph, and all graphs isomorphic to one of the graphs in the non-isomorphic list
def generateAllNonIsoCanonical(list_of_dict):
    
    #gets new dictionaries for permutations using one specific bond edge type (data)
    #also outputs the swapped versions only for that bond edge type (like it only gives a list of the "a")
    def allPermutationsForBondEdge(BDict, bEdgeIndex):
        #gets the key at index bEdgeIndex
        current_key = list(BDict.keys())[bEdgeIndex]
        current_value = BDict.get(current_key)
        shared_left_vertices = []
        shared_right_vertices = []
        permutation_length = len(current_value)
        for i in range(permutation_length):
            if current_value[i][0] not in [x[0] for x in shared_left_vertices]:
                left_indices_of_shared = []
                left_shared_vertex = current_value[i][0]
                for j in range(permutation_length):
                    if current_value[j][0] == left_shared_vertex:
                            left_indices_of_shared.append(j)
                if len(left_indices_of_shared) > 0:
                    shared_left_vertices.append([left_shared_vertex,left_indices_of_shared])

        for i in range(permutation_length):
            if current_value[i][1] not in [x[0] for x in shared_right_vertices]:
                right_indices_of_shared = []
                right_shared_vertex = current_value[i][1]
                for j in range(permutation_length):
                    if current_value[j][1] == right_shared_vertex:
                            right_indices_of_shared.append(j)
                if len(right_indices_of_shared) > 0:
                    shared_right_vertices.append([right_shared_vertex,right_indices_of_shared])

        #gets a valid permutation. eg: [0,1,2] or [3,4,5]
        def increment_permutation(permutation, max_number):
            if permutation == []:
                return []
            
            shortened_permutation = permutation[0:len(permutation)-1]

            if permutation[len(permutation)-1] >= max_number:
                incremented_short_permutation = increment_permutation(shortened_permutation, max_number - 1)
                next_incremented_value = -1
                if len(incremented_short_permutation) > 0:
                    next_incremented_value = incremented_short_permutation[len(incremented_short_permutation) - 1]
                return incremented_short_permutation + [next_incremented_value + 1]
            
            return shortened_permutation + [permutation[len(permutation) - 1] + 1]


        #gets all the permutations in the canonical ordering only. eg: 2 slots and max number 2 should give [0,1], [0,2],[1,2]
        def canonical_permutations(number_of_slots, max_number):
            final_list = []

            base_permutation = list(range(number_of_slots))
            final_list.append(base_permutation)
            incremented_permutation = copy.copy(base_permutation)
            incremented_permutation = increment_permutation(incremented_permutation, max_number)
            while incremented_permutation != base_permutation:
                final_list.append(incremented_permutation)
                incremented_permutation = increment_permutation(incremented_permutation, max_number)

            return final_list
        

# Further possible optimizations:
#
# generate whichever is a smaller set between left and right permutations first. then generate the other set one element a time, immediately throwing the element out if it is already in the first list.
# this could save on memory. and maybe both lists could be generated at the same time, comparing them to each other. there could be a way to do this that optimizes memory usage?
#
# our canonical_permutations function is recursive and slow. can we make it faster? there is probably a faster algorithm for generating that sort of thing

############################################################################################################ Calculating how many permutations there will be
        number_of_left_permutations = 1
        vertices_remaining = permutation_length
        for i in range(len(shared_left_vertices)):
            number_of_shared_vertices = len(shared_left_vertices[i][1])
            number_of_left_permutations *= binomial(vertices_remaining, number_of_shared_vertices)
            vertices_remaining -= number_of_shared_vertices
        number_of_left_permutations *= factorial(vertices_remaining)
        print(f'number of left permutations: {number_of_left_permutations}')

        number_of_right_permutations = 1
        vertices_remaining = permutation_length
        for i in range(len(shared_right_vertices)):
            number_of_shared_vertices = len(shared_right_vertices[i][1])
            number_of_right_permutations *= binomial(vertices_remaining, number_of_shared_vertices)
            vertices_remaining -= number_of_shared_vertices
        number_of_right_permutations *= factorial(vertices_remaining)
        print(f'number of right permutations: {number_of_right_permutations}')
############################################################################################################

        full_permutations = []
        full_left_permutations = []
        full_right_permutations = []

############################################################################################################ Left vertex permutations
        blank_permutation = [-1]*permutation_length
        full_left_permutations = [blank_permutation]
        max_number = permutation_length - 1
        for i in range(len(shared_left_vertices)):
            left_indices_of_shared = shared_left_vertices[i][1]
            old_permutations = full_left_permutations
            full_left_permutations = [] 

            for partial_permutation in old_permutations:
                for permutation_to_add in canonical_permutations(len(left_indices_of_shared), max_number):
                    indices_remaining = list(range(permutation_length))
                    new_permutation = copy.copy(partial_permutation)
                    for j in partial_permutation:
                        if j != -1:
                            indices_remaining.remove(j)
                    for j in range(len(permutation_to_add)):
                        new_permutation[left_indices_of_shared[j]] = indices_remaining[permutation_to_add[j]]
                    full_left_permutations.append(new_permutation)

            max_number -= len(left_indices_of_shared)

        empty_slots_remaining = 0
        empty_slot_indices = []
        for i in range(len(full_left_permutations[0])):
            if full_left_permutations[0][i] == -1:
                empty_slots_remaining += 1
                empty_slot_indices.append(i)
        remaining_permutations = Permutations(empty_slots_remaining)
        old_permutations = full_left_permutations
        full_left_permutations = [] 
        for partial_permutation in old_permutations:
            for permutation_to_add in remaining_permutations:
                indices_remaining = list(range(permutation_length))
                new_permutation = copy.copy(partial_permutation)
                for j in partial_permutation:
                    if j != -1:
                        indices_remaining.remove(j)
                for j in range(len(permutation_to_add)):
                    new_permutation[empty_slot_indices[j]] = indices_remaining[permutation_to_add[j] - 1]
                full_left_permutations.append(new_permutation)
############################################################################################################

############################################################################################################ Right vertex permutations
        blank_permutation = [-1]*permutation_length
        full_right_permutations = [blank_permutation]
        max_number = permutation_length - 1
        for i in range(len(shared_right_vertices)):
            right_indices_of_shared = shared_right_vertices[i][1]
            old_permutations = full_right_permutations
            full_right_permutations = [] 

            for partial_permutation in old_permutations:
                for permutation_to_add in canonical_permutations(len(right_indices_of_shared), max_number):
                    indices_remaining = list(range(permutation_length))
                    new_permutation = copy.copy(partial_permutation)
                    for j in partial_permutation:
                        if j != -1:
                            indices_remaining.remove(j)
                    for j in range(len(permutation_to_add)):
                        new_permutation[right_indices_of_shared[j]] = indices_remaining[permutation_to_add[j]]
                    full_right_permutations.append(new_permutation)

            max_number -= len(right_indices_of_shared)

        empty_slots_remaining = 0
        empty_slot_indices = []
        for i in range(len(full_right_permutations[0])):
            if full_right_permutations[0][i] == -1:
                empty_slots_remaining += 1
                empty_slot_indices.append(i)
        remaining_permutations = Permutations(empty_slots_remaining)
        old_permutations = full_right_permutations
        full_right_permutations = [] 
        for partial_permutation in old_permutations:
            for permutation_to_add in remaining_permutations:
                indices_remaining = list(range(permutation_length))
                new_permutation = copy.copy(partial_permutation)
                for j in partial_permutation:
                    if j != -1:
                        indices_remaining.remove(j)
                for j in range(len(permutation_to_add)):
                    new_permutation[empty_slot_indices[j]] = indices_remaining[permutation_to_add[j] - 1]
                full_right_permutations.append(new_permutation)
############################################################################################################

############################################################################################################ Mapping right vertex permutations to canonical left vertex permutations
        def right_to_left_canonical_permutation(right_permutation, shared_left_vertices):
            left_permutation = []
               
            # this actually generates the correspondig left permutation
            for i in range(len(right_permutation)):
                left_permutation.append(right_permutation.index(i))
            
            # this converts that supposed left permutation into its proper left permutation canonical form
            for i in range(len(shared_left_vertices)):
                indices = shared_left_vertices[i][1]
                sublist = []
                for j in range(len(left_permutation)):
                    if j in indices:
                        sublist.append(left_permutation[j])
                        left_permutation[j] = -1
                sublist = sorted(sublist)
                for j in range(len(left_permutation)):
                    if left_permutation[j] == -1:
                        left_permutation[j] = sublist[0]
                        sublist.pop(0)

            return left_permutation

        for i in range(len(full_right_permutations)):
            full_right_permutations[i] = right_to_left_canonical_permutation(full_right_permutations[i], shared_left_vertices)
############################################################################################################

############################################################################################################ Intersecting the sets of left vertex permutations and right vertex permutations
        full_permutations = [permutation for permutation in full_left_permutations if permutation in full_right_permutations]
############################################################################################################
        
        swapped_for_edge_type = [current_value]
                
        for i in range(0, len(full_permutations)): #go to length of permutation
            info = full_permutations[i]
            temp_values=[]
            for j in range(len(current_value)):
                ending = info[j]
                temp_values.append([current_value[j][0],current_value[ending][1]]) # This permutes right vertices (using 'left vertex' permutations)
            # #check if temp_values == current_value as sets (don't need this anymore because of our swap redundancy checks)
            # ordered_current_value = sorted(current_value)
            # if sorted(temp_values) != ordered_current_value:
            swapped_for_edge_type.append(temp_values)
        print(f'number of permutations after intersection: {len(full_permutations)}')
        #only change the values for the key we are working with (ie the bond edge type we are working with)
        return swapped_for_edge_type

    ## have to do the permutations first. but then you can do all number of multiple swaps for all bond edge types
    def allMultSwaps(BDict, all_possible_partial_dict):
        all_mult_swap_dicts = []
        all_possible_partial_dict_without_ratios = [t[0] for t in all_possible_partial_dict]
        pre_final = list(itertools.product(*all_possible_partial_dict_without_ratios))
        
        for i in range(len(pre_final)):
            final = dict.fromkeys(BDict[0])
            for j in range(len(final)):
                final[list(final.keys())[j]] = pre_final[i][j]
            all_mult_swap_dicts.append((final,BDict[1]))
        return all_mult_swap_dicts

    #get rid of duplicate dictionaries in the list of dictionaries
    #returns the distinct dictionaries
    def removeDupDict(list_of_dict):
        new_list = []
        for dictionary in list_of_dict:
            if dictionary not in new_list:
                new_list.append(dictionary)
        return new_list

    # Returns (non_iso_list, iso_list, canonical_list), lists of bond dictinoaries for all non-isomorphic graphs to one graph for one tile ratio, and all graphs isomorphic to one of the graphs in the non-isomorphic list, and all the canonical graphs for each non-iso graph
    def generateAllNonIsoCanonicalPartial(list_of_dict):
        non_iso_list = []
        iso_list = []
        canonical_list = []

        original_dict = list_of_dict[0]
        non_iso_list.append(original_dict)
        original_graph = graph_from_bdict(original_dict[0])
        canonical_list.append(original_graph.canonical_label(algorithm='sage'))

        for i in range (1, len(list_of_dict)):
            ratio = list_of_dict[i][1]
            current_dict = list_of_dict[i][0]
            current_graph = graph_from_bdict(current_dict)
            current_dict_can_label = current_graph.canonical_label(algorithm='sage')
            #if the current canonical label is not in the current list of canonical labels, that means it's not isomorphic
            if all([ current_dict_can_label != can_label for can_label in canonical_list ]):
                non_iso_list.append((current_dict, ratio))
                canonical_list.append(current_dict_can_label)
            else:
                #we should only find one thing in the non_iso_list that is isomorphic to the current graph anyway
                #because if we found more than one, that would be wrong. because non_iso_list should all have different canonical labels
                which_dict_index = [k for k, value in enumerate(canonical_list) if value == current_dict_can_label]
                iso_list.append((which_dict_index, current_dict, ratio))
        return (non_iso_list, iso_list, canonical_list)

    temp_non_iso_lists = []
    temp_canonical_lists = []

    for dictionary in list_of_dict:
        all_possible_partial_dict = []
        #gets the dictionaries for all the single swaps (each of the bond types)
        for j in range(len(dictionary[0])):
            all_possible_partial_dict.append((allPermutationsForBondEdge(dictionary[0],j), dictionary[1]))


        #gets the dictionaries for all multiple swaps (literally 1,2,3,etc. swaps, it gets all of them)
        complete_all_swaps = allMultSwaps(dictionary, all_possible_partial_dict)

        # #remove duplicates (not necessary if we remove all swap redundancies beforehand)
        # distinct_all_swaps = []
        # distinct_all_swaps = removeDupDict(complete_all_swaps)
        distinct_all_swaps = complete_all_swaps # this is for once we have removed all swap redundancies beforehand

        final_answer = ()
        if distinct_all_swaps != []:
            final_answer = generateAllNonIsoCanonicalPartial(distinct_all_swaps)
        else:
            final_answer = ([(dictionary[0],dictionary[1])], [], [graph_from_bdict(dictionary[0]).canonical_label(algorithm='sage')])
        temp_non_iso_lists.append(final_answer[0])
        temp_canonical_lists.append(final_answer[2])
    
    non_iso_list = temp_non_iso_lists[0]
    iso_list = []
    canonical_list = temp_canonical_lists[0]

    for i in range(1, len(temp_canonical_lists)):
        for j in range(len(temp_canonical_lists[i])):
            canonical_graph = temp_canonical_lists[i][j]
            if all([ canonical_graph != x for x in canonical_list ]):
                non_iso_list.append(temp_non_iso_lists[i][j])
                canonical_list.append(canonical_graph)
            else:
                which_dict_index = [k for k, value in enumerate(canonical_list) if value == canonical_graph]
                iso_list.append((which_dict_index, temp_non_iso_lists[i][j][0], temp_non_iso_lists[i][j][1]))

    return (non_iso_list, iso_list)

list_of_bond_dictionaries = []
list_of_labels = []
#Construct and plot one graph for each possible multiplicity
ordering_to_use = degrees_diversities_ordering
for i,multiplicities in enumerate(possible_multiplicities):
    bond_dictionary_and_labels = construct_bond_dictionary(tiles, multiplicities, ordering_to_use, avoid_multiple_edges)
    bond_dictionary = bond_dictionary_and_labels[0]
    ratio = bond_dictionary_and_labels[1]
    labels = bond_dictionary_and_labels[2]
    print(f"Final bond dictionary {i+1}:")
    print(bond_dictionary)
    list_of_bond_dictionaries.append((bond_dictionary, ratio))
    list_of_labels.append(labels)

ANSWER = generateAllNonIsoCanonical(list_of_bond_dictionaries)
#print(ANSWER)
print(f'\nnumber of non-isomorphic graphs: {len(ANSWER[0])}\n')

#This list stores all the graphs for later visualization
list_of_graphs = []

for j,i in enumerate(ANSWER[0]):
    ratio = i[1]
    print(f"Graph {j+1}:")
    print(f"bond dictionary: {i[0]}")
    print(f"ratio: {ratio}")
    graph = labeled_graph_from_bond_dictionary(i[0], list_of_labels[possible_multiplicities.index(ratio)])
    P = graph.plot(layout='spring',save_pos=True,iterations=1000,edge_labels=True)
    list_of_graphs.append(graph)
    P.show()
```
